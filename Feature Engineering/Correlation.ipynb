```python
import pandas as pd

df = pd.read_csv('home_prices.csv')
df.head()
```

```python
df.color.unique()
```

```python
df_encoded = pd.get_dummies(df, columns=['color'], drop_first=True)
df_encoded.head()
```

```python
cm = df_encoded.corr()
cm
```

```python
cm_abs = abs(cm['price_lakhs'])
```

```python
threshold = 0.2
cm_abs
```

```python
cm_price = cm_abs > threshold
```

```python
selected_features = cm_price[cm_price].index.drop('price_lakhs')
selected_features
```

```python
X = df[selected_features]
fullX = df_encoded.drop('price_lakhs', axis=1)
y = df['price_lakhs']
```

```python
from sklearn.model_selection import train_test_split

X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
fullX_training, fullX_test, fully_training, fully_test = train_test_split(fullX, y, test_size=0.2, random_state=5)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model_X = LinearRegression()
```

```python
model.fit(X_training, y_training)
model_X.fit(fullX_training, fully_training)
```

```python
y_pred = model.predict(X_test)
fully_pred = model_X.predict(fullX_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_pred, y_test)
r2 = r2_score(y_pred, y_test)

mse, r2
```

```python
full_mse = mean_squared_error(fully_pred, fully_test)
full_r2 = r2_score(fully_pred, fully_test)

full_mse, full_r2
```
